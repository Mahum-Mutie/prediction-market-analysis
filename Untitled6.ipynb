```python
import ast

def parse_prices(x):
    try:
      return [float(i) for i in ast.literal_eval(x)]
    except:
      return None

df['parced_prices'] = df['outcome_prices'].apply(parse_prices)
df[['question','outcome_prices','parced_prices']].head()
```

```python
df['confidence'] = df['parced_prices'].apply(
    lambda x: max(x) if isinstance (x, list) else None
)

df[['question', 'parced_prices','confidence']].head(10)
```

```python
from collections import Counter
import re

other_questions = df[df['category'] == 'other']['question'].dropna()

stopwords = {'will', 'the', 'in', 'on', 'by', 'of', 'a', 'an', 'to', 'for', 'and', 'or', 'at', 'be', 'is', 'are', 'with', 'before', 'after', 'during', 'more', 'than', 'under', 'any', 'have', 'not'}

words = []

for q in other_questions:
  q = q.lower()
  q = re.sub(r'[^ɑ\w\s$]','', q)
  for word in q.split():
    if word not in stopwords and len(word)>2:
      words.append(word)
Counter(words).most_common(40)
```

```python
def categorized(q):
  q = q.lower()

 #Politics
  if any(word in q for word in ['trump', 'election', 'kamala', 'rally', 'congress','Biden', 'Obama', 'president', 'ceasefire', 'war', 'iran', 'isreal', 'country', 'shutdown', 'goverment', 'mh370', 'dnc', 'debate', 'democratic', 'state', 'minister', 'announce', 'announced', 'found', 'discovered', 'event', 'happended', ]):
    return 'politics'

  # Economics/Finance
  elif any(word in q for word in ['fed', 'rate','inflation','gold', 'market', 'stock', 'cap', 'gme', '$', 'ecomony']):
    return 'economics'

  # Sports
  elif any(word in q for word in ['win', 'vs', 'beat', 'points', 'match', 'fight', 'final', 'tournament', 'nba', 'nfl', 'soccer', 'masters', 'ballon', 'team', 'super bowl', 'falcons', 'vikings', 'patriots', 'eagles', 'celtics', 'player', 'score', 'coach']):
    return 'Sports'

  # Crypto
  elif any(word in q for word in ['bitcoin', 'ethereum', 'crypto', 'cryptocurrency','eth', 'sol']):
    return 'crypto'

  # Entertainment
  elif any(word in q for word in ['movie', 'opening', 'panda', 'box office', 'panda', 'film', 'music']):
    return 'entertainment'

  #Tech
  elif any(word in q for word in ['elon', 'tweet', 'twitter', 'ai', 'tech', 'gpt', 'ai', 'tiktok', 'spacex']):
    return 'tech'

  #Legal
  elif any(word in q for word in ['guilty', 'sentence', 'jail', 'trail', 'prison']):
    return 'legal'


  return 'other'
```

```python
df['category'] = df['question'].apply(categorized)
category_volume = df.groupby('category')['volume'].mean().sort_values(ascending=False)
print(category_volume)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
category_volume.plot(kind='bar')
plt.title('Average Trading Volume by Question Category')
plt.xlabel('Category')
plt.ylabel('Average Volume')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
category_counts = df['category'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Question Categories')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
